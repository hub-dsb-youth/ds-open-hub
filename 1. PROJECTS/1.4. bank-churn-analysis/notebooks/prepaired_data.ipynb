```python
# import thư viện 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = pd.read_csv("data/raw/Bank-Customer-churn.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
data_clean =  data.drop_duplicates(subset="CustomerId")
```

```python
data_clean.info()
```

```python
data_clean.describe()
```

```python
ax = sns.boxplot(data = data_clean, x = "Age",  y = "Balance", palette= "colorblind")
```

```python
# Xử lý cho age
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25) # Phân vị thứ nhất
    Q3 = df[column].quantile(0.75) # Phân vị thứ ba 
    IQR = Q3 - Q1 # Công thức IQR 
    lower = Q1 - 1.5 * IQR # Outlier under
    upper = Q3 + 1.5 * IQR # Outlier top 
    return df[(df[column] >= lower) & (df[column] <= upper)]

data_remove = remove_outliers(data_clean, 'Age')
```

```python
data_remove.describe()
```

```python
# Xử lý cho Balance
lower = data_remove['Balance'].quantile(0.01)
upper = data_remove['Balance'].quantile(0.99)
print(f"Lower: {lower}")
print(f"upper: {upper}")

data_remove = data_remove[
    (data_remove['Balance'] >= lower) &
    (data_remove['Balance'] <= upper)
]
```

```python
data_remove.describe()
```

```python
data_remove.info()
```

```python
# Age group  <30 => young, < 50 "Middle" , >=50 "Old"
def age_group(age):
    if age < 30:
        return "Young"
    elif 30 <= age < 50:
        return "Middle"
    else:
        return "Old"
data_age_group = data_remove.copy()
data_age_group["Age Group"] = data_age_group["Age"].apply(age_group)
data_age_group.head(10)
```

```python
# Balance Category (Low: <50k, Medium: 50k-100k, High: >100k).
def balance (balance):
    if balance < 50000:
        return "Low"
    elif 50000 <= balance <= 100000:
        return "Medium"
    else:
        return "High"
data_balance = data_age_group.copy()
data_balance["Balance Category"] = data_balance["Balance"].apply(balance)
data_balance.head(10)
```

```python
data_balance.to_csv("data/processed/Bank-Customer-churn-clean.csv")
print("Đã lưu file thành công")
```
